## **Data Preparation**

```python
df = df.dropna()
```

```python
# Transformando os dados para o formato adequado

from pyspark.sql.functions import col
from pyspark.sql.types import IntegerType, DateType, StringType

df = df.withColumn("Vlr_Bruto", col("Vlr_Bruto").cast(IntegerType())) \
       .withColumn("Vlr_Liquido", col("Vlr_Liquido").cast(IntegerType())) \
       .withColumn("Data_x", col("Data_x").cast(DateType())) \
       .withColumn("Data_y", col("Data_y").cast(DateType()))
```

```python
#Convertendo Feedback em valores numéricos
from pyspark.sql.functions import when
df = df.withColumn("Sentimento", when(df["Sentimento"] == "Feedback Positivo", 1)
                               .when(df["Sentimento"] == "Feedback Negativo", 0)
                               .otherwise(df["Sentimento"]))
```

```python
df = df.withColumn("Sentimento", df["Sentimento"].cast(IntegerType()))
```

```python
# Analisando se existe correlação entre duas colunas específicas

correlation = df.stat.corr("Vlr_Liquido", "Sentimento")
print(f"Correlação entre Vlr_Liquido e Sentimento: {correlation}")
```

```python
import matplotlib.pyplot as plt

# Coletar os dados das duas colunas em listas
coluna1_values = df.select("Vlr_Liquido").rdd.flatMap(lambda x: x).collect()
coluna2_values = df.select("Sentimento").rdd.flatMap(lambda x: x).collect()

# Criar o gráfico de dispersão
plt.figure(figsize=(10, 6))
plt.scatter(coluna1_values, coluna2_values)
plt.title(f"Scatter Plot entre Vlr_Liquido e Sentimento (Correlação = {correlation:.2f})")
plt.xlabel("Sentimento")
plt.ylabel("Vlr_Liquido")
plt.show()
```

```python
#Filtrando um intervalo de datas

from pyspark.sql.functions import col, datediff, to_date

df = df.withColumn("Data_x", to_date(col("Data_x")))
df = df.withColumn("Data_y", to_date(col("Data_y")))
```

```python
df_filtered = df.filter(datediff(col("Data_x"), col("Data_y")) <= 60)
df_filtered = df_filtered.filter(datediff(col("Data_x"), col("Data_y")) >= 0)
```

```python
df_filtered = df_filtered.select("Sentimento", "Vlr_Liquido")
```

```python
# Definindo as features

from pyspark.ml.feature import VectorAssembler

assembler = VectorAssembler(
    inputCols=["Sentimento"],
    outputCol="features"
)

df_transformed = assembler.transform(df_filtered)
```

```python
# Base de treino e teste

train_data, test_data = df_transformed.randomSplit([0.8, 0.2], seed=1234)
```

```python
# Treinando o modelo

from pyspark.ml.regression import RandomForestRegressor

rf = RandomForestRegressor(featuresCol="features", labelCol="Vlr_Liquido")
rf_model = rf.fit(train_data)
```

```python
predictions = rf_model.transform(test_data)
```

```python
# Avaliando o modelo

from pyspark.ml.evaluation import RegressionEvaluator

evaluator = RegressionEvaluator(
    labelCol="Vlr_Liquido",
    predictionCol="prediction",
    metricName="rmse"
)

rmse = evaluator.evaluate(predictions)
print(f"Root Mean Squared Error (RMSE) no conjunto de teste: {rmse}")
```

```python
feature_importances = rf_model.featureImportances
print(f"Importância das features: {feature_importances}")
```
